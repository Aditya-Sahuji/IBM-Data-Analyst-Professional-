Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
Query2 = """
SELECT ConvertedComp 
FROM master
"""

ConvertedComp = pd.read_sql_query(Query2, conn)
ConvertedComp.plot(kind='hist', figsize=(8, 5))
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
Query2 = """
SELECT Age
FROM master
"""

Age = pd.read_sql_query(Query2, conn)
Age.plot(kind='box', figsize=(8, 10))
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
Query3 = """
SELECT Age,WorkWeekHrs
FROM master
"""

Scatter = pd.read_sql_query(Query3, conn)
Scatter.plot(kind='scatter', x='WorkWeekHrs', y='Age', figsize=(10, 6))
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
Query3 = """
SELECT Age,WorkWeekHrs,CodeRevHrs
FROM master
"""

bubble = pd.read_sql_query(Query3, conn)
bubble.plot(kind='scatter',
                    x='CodeRevHrs',
                    y='WorkWeekHrs',
                    figsize=(10, 8),
                    s= Age,  # pass in weights
                    )
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
Query4 = """
SELECT DatabaseDesireNextYear, count(*) as count
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by count desc limit 5
"""

DatabaseDesireNextYear= pd.read_sql_query(Query4, conn)
x = DatabaseDesireNextYear['count']
label = DatabaseDesireNextYear['DatabaseDesireNextYear']
plt.pie(x, labels=label, autopct='%1.1f%%',radius=1.5)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
Query5 = """
SELECT Avg(WorkWeekHrs), Avg(CodeRevHrs), Age 
FROM master 
where Age between 30 and 35
group by Age
"""
stacked = pd.read_sql_query(Query5, conn)
WorkWeekHrs= stacked['Avg(WorkWeekHrs)']
CodeRevHrs= stacked['Avg(CodeRevHrs)']
Age= stacked['Age']
fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(Age, WorkWeekHrs, label='WorkWeekHrs')
ax.bar(Age, CodeRevHrs, bottom=WorkWeekHrs,
       label='CodeRevHrs')

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
Query8= """
SELECT Avg(ConvertedComp), Age
FROM master
where age between 45 and 60
group by Age
"""
line = pd.read_sql_query(Query8, conn)

line.plot(kind='line' ,figsize=(8, 5))
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
Query7 = """
SELECT MainBranch, count(*)
FROM master
group by MainBranch
"""
BARH = pd.read_sql_query(Query7, conn)
BARH.plot(kind='barh', figsize=(8, 5))


QUERY = """
SELECT respondent
FROM master

"""
pd.read_sql_query(QUERY,conn)
```
